```python
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
movies_genres = movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count',
        'vote_average','profit']]
movies_genres.head()
```

```python
from pandas import Series, DataFrame

split = movies_genres['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name='genres_split'
del movies_genres['genres']
movies_genres = movies_genres.join(split)
```

```python
movies_genres.head()
```

```python
#1. Which genres are the most common?
```

```python
genres_count = pd.DataFrame(movies_genres.groupby('genres_split').original_title.nunique()).sort_values('original_title',ascending= True)
genres_count
```

```python
genres_count['original_title'].plot.pie(title = 'Movies per Genre in %', autopct='%1.1f%%',figsize =(10,10))
```

```python
genres_count['original_title'].plot.barh(title = 'Movies per Genre', color ='green', figsize =(10,10))
```

```python
#2. Which genres have highest avg budget and revenue?
```

```python
genres_avg = movies_genres.groupby('genres_split').mean(numeric_only=True)
pd.options.display.float_format = '{:2f}'.format
genres_avg.sort_values('budget', ascending= True, inplace= True)
genres_avg
```

```python
genres_avg[['budget','revenue']].plot.barh(title = 'Budget and Revenue by Genre', color={'Darkblue','c'}, figsize =(10,10))
```

```python
#2.5 Which genres have highest average Profit?
```

```python
genres_avg.sort_values('profit', ascending= False, inplace= True)
```

```python
genres_avg['profit'].plot.barh(title = 'Profit by Genre', color={'Darkblue'}, figsize =(10,9))
```

4. A look at profit per genre per year

```python
# replacing release_date with release_year

movies_genres = movies[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count',
        'vote_average','profit']]
```

```python
split = movies_genres['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name='genres_split'
del movies_genres['genres']
movies_genres = movies_genres.join(split)
```

```python
movies_genres.head()
```

```python
time_genre = pd.DataFrame(movies_genres.groupby(['release_year','genres_split'])['profit'].mean(numeric_only=True))
time_genre
```

```python
final_genre = pd.pivot_table(time_genre, values= 'profit', index= 'genres_split', columns='release_year')
final_genre
```

```python
import matplotlib.pyplot as plt
sns.set(rc = {'figure.figsize':(15,10)})
sns.heatmap(final_genre, cmap='YlGnBu', linewidths = .5)
plt.title('Genres by profit per year')
```
